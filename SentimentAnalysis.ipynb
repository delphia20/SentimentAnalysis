```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
```

```python
reviews = {
    "Review": [
        "This product is amazing",
        "Very good quality",
        "I love this item",
        "Worst product ever",
        "Very disappointed",
        "Not worth the money",
        "Average product",
        "Good but expensive",
        "Excellent quality",
        "Terrible experience"
    ]
}

df = pd.DataFrame(reviews)

df.head()
```

```python
def get_sentiment(text):
    
    polarity = TextBlob(text).sentiment.polarity
    
    if polarity > 0:
        return "Positive"
    
    elif polarity < 0:
        return "Negative"
    
    else:
        return "Neutral"
```

```python
df["Sentiment"] = df["Review"].apply(get_sentiment)

df
```

```python
df["Sentiment"].value_counts()
```

```python
sentiment_counts = df["Sentiment"].value_counts()

plt.figure(figsize=(6,4))

sentiment_counts.plot(kind="bar")

plt.title("Sentiment Analysis Results")

plt.xlabel("Sentiment")

plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(6,6))

sentiment_counts.plot(kind="pie", autopct="%1.1f%%")

plt.ylabel("")

plt.title("Sentiment Distribution")

plt.show()
```

# Sentiment Analysis Insights

1. Reviews were classified into Positive, Negative and Neutral categories.
2. TextBlob was used to calculate sentiment polarity.
3. Positive reviews dominated the dataset.
4. Negative reviews highlighted customer dissatisfaction.
5. Sentiment analysis helps businesses understand customer opinions.
